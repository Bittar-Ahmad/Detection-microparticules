## Particle simulation

```python
import matplotlib.pylab as plt
import numpy as np
import numpy.random
import scipy
import skimage
import pymrt as mrt
import raster_geometry as rg
import time
```

```python
# constants
def setParameters(shapeimg=60,# size of the image
                  radius=8, # radius of HR ball
                  step=0.1, # relative to 1
                  slices=7,# number of 2D samples
                  zoomfactor=0.1,
                  approxorder=1):
    return(shapeimg,radius,step,slices,zoomfactor,approxorder)

# globals, avoid using them
shapeimg,radius,step,slices,zoomfactor,approxorder = setParameters()
```

```python
## Random placement
np.random.seed(42)
center=np.random.normal(loc=0.5,scale=step,size=3)



# Parameters

# Dimension of the 3D image
Nx,Ny,Nz = (shapeimg,shapeimg,shapeimg)



# Particle simulation
# perfect sphere
particle =rg.sphere((Nx,Ny,Nz),radius,position=center).astype(float)
print("center=",center*Nx)
print(particle.shape)





########################################### Noise ###################################
#Adding noise
#Ajouter de bruits 
#Une fonction noise Gaussian
def noisy_particle_Guassian(particle, mean, variance):    
    # Generate Gaussian noise
    Nx, Ny, Nz= particle.shape
    sigma = np.sqrt(variance)
    gauss = np.random.normal(mean,sigma,(Nx, Ny, Nz))
    gauss = gauss.reshape(Nx, Nz, Nz)
    noisy_particle = particle + gauss
    return noisy_particle 

def noisy_particle_SaltPepper(particle, s_vs_p, amount):    
      Nx, Ny, Nz = particle.shape
      out = np.copy(particle)
      
      # Mode de Salt
      num_salt = np.ceil(amount * particle.size * s_vs_p)
      coords = [np.random.randint(0, i - 1, int(num_salt))
              for i in particle.shape]
      out[coords] = 1

      # Pepper mode
      num_pepper = np.ceil(amount* particle.size * (1. - s_vs_p))
      coords = [np.random.randint(0, i - 1, int(num_pepper))
              for i in particle.shape]
      out[coords] = 0
      return out
    
def noisy_particle_Poisson(particle, param):    
    vals = len(np.unique(particle))
    vals = param ** np.ceil(np.log2(vals))
    noisy_particle = np.random.poisson(particle * vals) / float(vals)
    return noisy_particle


#Noisy particle Gaussian
#particle = noisy_particle_Guassian(particle, 0, 0.0001)

#Noisy particle Salt&Pepper
#particle = noisy_particle_SaltPepper(particle, 0.01, 0.04)

#Noisy particle Poisson
noisyparticle = noisy_particle_Poisson(particle, 500)
##########################################################################################

plt.figure()
plt.imshow(particle[:,:,int(Nz*center[2])])
plt.scatter([Nx*center[1]],[Ny*center[0]], marker="x", )
plt.show()

plt.figure()
plt.imshow(noisyparticle[:,:,int(Nz*center[2])])
plt.scatter([Nx*center[1]],[Ny*center[0]], marker="x", )
plt.show()


# Diagonal of the frames
d = np.sqrt(Nx**2+Ny**2)

# z-coordinates of the frames
#frames  = np.arange(0,nz) # if zoomed
frames  = np.arange(0,Nz)
# Number of frames
m = len(frames)

```

```python
## Moving particles: random walk (Gaussian)

# Small movements:
sigma=0.003 # the values are between 0 and 1 (proportion of the total image)
moves = np.random.normal(0, sigma, (Nz,3))


moving_particle =rg.sphere((Nx,Ny,Nz),radius,position=center).astype(float)
newcenter=center
plt.figure()
#plt.imshow(moving_particle[:,:,int(Nz*center[2])])
for i in range(Nz):
    plt.scatter([Nx*newcenter[1]],[Ny*newcenter[0]], marker="x", color = 'red')
    newcenter = newcenter + moves[i]
    #print(Nx*newcenter)
    moving_particle[:,:,i] = rg.sphere((Nx,Ny,Nz),radius,position=newcenter).astype(float)[:,:,i]
    

plt.show()  # all center locations
```

# Preprocessing

## Variation matrix

```python
# Total variation 3D matrix of the image frames
def W_mat(image, frames):
    nx,ny,nz = image.shape
    nz = len(frames)
    W = np.zeros((nx,ny,nz))
    for i in range(nx):
        for j in range(ny):
            k_translated = 0
            for k in frames:
                varX = 0
                varY = 0
                if i==0:
                    varX += (image[i+1,j,k]-image[i,j,k])**2
                elif i==nx-1:
                    varX += (image[i-1,j,k]-image[i,j,k])**2
                else:
                    varX += (image[i-1,j,k]-image[i,j,k])**2+(image[i+1,j,k]-image[i,j,k])**2
                if j==0:
                    varY += (image[i,j+1,k]-image[i,j,k])**2
                elif j==ny-1:
                    varY += (image[i,j-1,k]-image[i,j,k])**2
                else:
                    varY += (image[i,j-1,k]-image[i,j,k])**2+(image[i,j+1,k]-image[i,j,k])**2
                W[i,j,k_translated] = np.sqrt(varX+varY)
                k_translated +=1
    return W

#normal
plt.figure(figsize=(15,10))
plt.imshow(W_mat(particle,frames)[:,:,int(Nz*center[2])])

#noisy
plt.figure(figsize=(15,10))
plt.imshow(W_mat(noisyparticle,frames)[:,:,int(Nz*center[2])])
```

```python
# Total variation 3D matrix of the image frames with edge DETECTION
def W_mat_Edge(image, frames):
    nx,ny,nz = image.shape
    nz = len(frames)
    W = np.zeros((nx,ny,nz))
    for i in range(nx):
        for j in range(ny):
            k_translated = 0
            for k in frames:
                varX = 0
                varY = 0
                if i==0:
                    varX += (image[i+1,j,k]-image[i,j,k])**2
                elif i==nx-1:
                    varX += (image[i-1,j,k]-image[i,j,k])**2
                else:
                    varX += (image[i-1,j,k]-image[i,j,k])**2+(image[i+1,j,k]-image[i,j,k])**2
                if j==0:
                    varY += (image[i,j+1,k]-image[i,j,k])**2
                elif j==ny-1:
                    varY += (image[i,j-1,k]-image[i,j,k])**2
                else:
                    varY += (image[i,j-1,k]-image[i,j,k])**2+(image[i,j+1,k]-image[i,j,k])**2
                W[i,j,k_translated] = np.sqrt(varX+varY)
                k_translated +=1
                
    
       
    ############ Edge detection with convolve2d:  ####################    
    from scipy import signal

    grad = np.zeros((nx,ny,nz))    
    
    #Seems the best among the other kernels
    edge_kernel1 = np.array([[-1, -1, -1],
                       [-1, 8, -1],
                       [-1, -1, -1]])
    
    #not good at all
    edge_kernel2 = np.array([[1, 0, -1],
                       [0, 0, 0],
                       [-1, 0, 1]])
    
    #Has close results than edge_kernel1, but the latter is better.
    edge_kernel3 = np.array([[0, -1, 0],
                       [-1, 4, -1],
                       [0, -1, 0]])
    
    #To ignore this kernel
    scharr = np.array([[ -3-3j, 0-10j,  +3 -3j],
                   [-10+0j, 0+ 0j, +10 +0j],
                   [ -3+3j, 0+10j,  +3 +3j]]) # Gx + j*Gy
    
    # ==> choose edge_kernel1:
    for zz in range(nz):        
        ascent = W[:,:,zz]
        #grad[:,:,zz] = signal.convolve2d(ascent, scharr, boundary='symm', mode='same')
        grad[:,:,zz] = signal.convolve2d(ascent, edge_kernel1, boundary='symm', mode='same')
        
        #Uncomment the following two lines to obtain an approximation of the original W before edge detection
        #grad[grad > 255] = 255
        #grad[grad < 0] = 0
    return grad
##############################################################################

# normal
plt.figure(figsize=(15,10))
plt.imshow(W_mat_Edge(particle,frames)[:,:,int(Nz*center[2])])


# noisy
plt.figure(figsize=(15,10))
plt.imshow(W_mat_Edge(noisyparticle,frames)[:,:,int(Nz*center[2])])
```

Both variation matrices are robust to noise

```python
# Keeping only frames intersecting the sphere

def intersecting_frames(image,frames):
    start = -1
    end = -1
    index = 0
    m = len(frames)
    while end<0 and index<m:
        M = np.amax(image[:,:,frames[index]])
        if M>0 and start<0:
            start = index
        if start >= 0 and M==0:
            end = index
        index+=1
    if end<0:
        end = m
    return frames[start:end]




# normal
framesInter = intersecting_frames(particle, frames)
print(framesInter)
mInter = len(framesInter)
print(mInter)

#noisy (be carefull that changing the noise does not affect this step)
noisyframesInter = intersecting_frames(noisyparticle, frames)
print(noisyframesInter)
noisymInter = len(noisyframesInter)
print(noisymInter)
```

# Optimization problem

```python
# Objective function
def objective_function(image,frames):
    nx,ny,nz = image.shape
    m = len(frames)
    W = W_mat(image,frames)
    def obj(alpha,beta,gamma,xi,eta):
        res = 0
        for k in range(m):
            for i in range(nx):
                for j in range(ny):
                    res += W[i,j,frames[k]-frames[0]]*(-2*i*alpha -2*j*beta + xi - eta[k] +i**2+j**2)**2
        return res/2
    return obj


# Taking the intersecting frames should not changes the score
obj = objective_function(particle,frames)
print(obj(10,10,10,10,np.ones(m)))

obj = objective_function(particle,framesInter)
print(obj(10,10,10,10,np.ones(m)))
```

## Fix particles

### Matrices

```python
def Create_A_and_b(W):
    "Cette fonction permet de creer la matrice A de taille n * 4+m et le vecteur b de taille n"
    
    #Convertir la matrice W en 2D -- Taille: n * s(3) avec n = s(1)*s(2)
    s1 = np.size(W,0)
    s2 = np.size(W,1)
    s3 = np.size(W,2)
    
    W_2D = W.reshape(s1*s2, s3)

#Definir le vecteur f_1 de taille 4+m (vecteur de 0 sauf le 1er element est 1)    
    f_1 = np.zeros((4+s3,1))
    f_1[0] = 1   
    
#Definir le vecteur f_2 de taille 4+m (vecteur de 0 sauf le 2eme element est 1)
    f_2 = np.zeros((4+s3,1))
    f_2[1] = 1    
    
#Definir le vecteur f_3 de taille 4+m (vecteur de 0 sauf le 3eme element est 1)
    f_3 = np.zeros((4+s3,1))
    f_3[3] = 1    


    
# Definir les vecteurs x et y de taille n chacun
    x = np.zeros((s1*s2,1))
    y = np.zeros((s1*s2,1))
    for jj in range(s1*s2):
        x[jj] = np.floor((jj)/s1)
        y[jj] = np.mod(jj,s1)

        
#Definir un vecteur unite de taille n
    vecteur_unitaire = np.ones((s1*s2,1));    
        
    A = np.zeros((np.size(W_2D, 0), 4+s3, s3))
    b = np.zeros((np.size(W_2D, 0), s3));
    
    listA = [] #empty listA
    listb = []  #empty listb
    
    for kk in range(s3):
        #Definir le vecteur f_{4+i} de taille 4+m (vecteur de 0 sauf le (4+i)th element est 1)
        f_4_i = np.zeros((4+s3,1))   
        f_4_i[4+kk] = 1
        A[:, :, kk] = np.diag(np.sqrt(W_2D[:, kk])) .dot(-2*np.kron(np.transpose(f_1), x) - 2*np.kron(np.transpose(f_2), y) + np.kron(np.transpose(f_3), vecteur_unitaire) - np.kron(np.transpose(f_4_i), vecteur_unitaire))
        listA.append(A[:, :, kk])
        
        b[:, kk] = np.multiply(-np.sqrt(W_2D[:, kk]), np.transpose(x**2)) + np.multiply(-np.sqrt(W_2D[:, kk]), np.transpose(y**2))
        listb.append(b[:, kk])
        

    return  listA, listb

```

### Interior points with Newton's method

```python
import numpy as np
import matplotlib.pyplot as plt

# Centering step

# fonction à minimiser
def fct(listA, listb):
    m = len(listA)
    def f(v): 
        return np.sum([np.linalg.norm(listA[i]@v - listb[i])**2 for i in range(m)])/2
    return f

# v = np.concatenate((alpha, beta, gamma, xsi, eta))


# Fonction avec barrière:
def fct_bar(listA, listb, z, R, d, t): # t est le paramètre de barrière
    m = len(listA)
    def f(v):
        constr1 = v[4:]
        constr2 = -(v[4:] + (v[2]-z)**2 - R**2)
        constr3 = [-(v[0]**2 + v[1]**2 - v[3])]
        constr4 = [-(v[3] - d**2)]
        constr = np.concatenate((constr1,constr2,constr3,constr4))
        if np.min(constr) <= 0:
            #raise NameError("v non admissible")
            return np.inf
        else:
            return np.sum([np.linalg.norm(listA[i]@v - listb[i])**2 for i in range(m)])/2 - (1/t)*np.sum(np.log(constr))
    return f



# Gradient en v de la fonction à minimiser
def grad_f(listA, listb, z, R, d, t, v):
    n = len(v)
    m = len(listA)
    res = np.sum([listA[i].T@(listA[i]@v - listb[i]) for i in range(m)], axis=0)
    bar = np.zeros(n)
    # terme dû à la fonction barrière pour constr3 et constr4
    x = (-(v[0]**2 + v[1]**2 - v[3]))
    bar[0] += -2*v[0]/x
    bar[1] += -2*v[1]/x
    bar[3] += 1/x - 1/(d**2-v[3])
    for j in range(4,n):
        u = v[j]
        # terme dû à la fonction barrière pour constr1 et constr2
        x = R**2 - u - (v[2]-z[j-4])**2
        bar[2] += -2*(v[2]-z[j-4])/x
        bar[j] += 1/u - 1/x  
    return res - bar/t



# Hessienne
def hess_f(listA, listb, z, R, d, t, v):
    n = len(v)
    m = len(listA)
    res = np.sum([listA[i].T@listA[i] for i in range(m)], axis=0)
    bar = np.zeros((n,n))
    x = (v[0]**2 + v[1]**2 - v[3])**2
    bar[0,0] += (-2*(v[3] - v[0]**2 - v[1]**2) - 4*v[0]**2)/x
    bar[1,1] += (-2*(v[3] - v[0]**2 - v[1]**2) - 4*v[1]**2)/x
    bar[3,3] += -1/x - 1/(v[3]-d**2)**2
    b = -4*v[0]*v[1]/x
    bar[0,1] += b
    bar[1,0] += b
    b = 2*v[0]/x
    bar[0,3] += b
    bar[3,0] += b
    b = 2*v[1]/x
    bar[1,3] += b
    bar[3,1] += b
    for j in range(4,n):
        u = v[j]
        # terme dû à la fonction barrière pour constr1 et constr2
        x = (R**2 - u - (v[2]-z[j-4])**2)**2
        bar[2,2] += (-2*(R**2 - u - (v[2]-z[j-4])**2) - 4*(v[2]-z[j-4])**2)/x
        bar[j,j] += -1/u**2 - 1/x
        b = 2*(v[2]-z[j-4])/x
        bar[2,j] += b
        bar[j,2] += b
    return  res - bar/t





# Backtracking line search
def lineSearch(f,x,delta,grad,alpha,beta,t=1): # 0<alpha<1/2 et 0<beta<1
    while f(x+t*delta) >= f(x) + alpha*t*(grad.T@delta) :
        t = beta*t
    return t

def graddesc(f,v,grad, t=1000):
    crit = True
    while crit:
        if f(v - t*grad) < f(v):
            v = v - t*grad
            crit = False
        t = t/2
    return v

# Centering step
# On retourne la liste des points et des decr pour les graphes à la fin
def centering_step(listA, listb, z, R, d, t, v0, crit, alpha = 0.4, beta = 0.9): # t et eps > 0
    # méthode de Newton
    res = [v0]
    v = v0
    decr = np.inf
    liste_decr=[]
    f = fct_bar(listA, listb, z, R, d, t)
    grad = grad_f(listA, listb, z, R, d, t, v)
    hess = hess_f(listA, listb, z, R, d, t, v)
    hess_inv = np.linalg.inv(hess)
    delta = -hess_inv@grad
    decr = grad.T@hess_inv@grad
    #print("decr = ",decr)
    #print(np.linalg.norm(grad))
    #while np.linalg.norm(grad) > crit:
    while decr/2 >= crit:
        v = v + lineSearch(f,v,delta,grad,alpha,beta)*delta
        #v = graddesc(f,v,grad)
        res.append(v)
        grad = grad_f(listA, listb, z, R, d, t, v)
        hess = hess_f(listA, listb, z, R, d, t, v)
        hess_inv = np.linalg.inv(hess)
        delta = -hess_inv@grad
        decr = grad.T@hess_inv@grad
        liste_decr.append(decr)
        print("decr = ",decr)
        #print(delta)
        #print("v_cent = ",v)
        #print("f_bar(v_cent) = ",f(v))
    return (res,liste_decr)


# Barrier method
# On retourne la liste des points et des decr
def barr_method(listA, listb, z, R, d, v0,eps, crit, alpha =0.4, beta = 0.9, t0 = 1, mu = 10): # t0, mu et eps > 0
    start = time.time()
    res = [v0]
    liste_decr = []
    n = len(v0)
    v = v0
    t = t0
    while n/t >= eps:
        #print("v = ",v)
        print("f(v) = ",fct(listA,listb)(v), " Barrier parameter = ", t)
        (v, liste_decr_centering) = centering_step(listA, listb, z, R, d, t, v,crit, alpha, beta)
        liste_decr+= liste_decr_centering
        t = mu*t
        res = res + v
        v = v[-1]
    print("Computation time: ", time.time()-start, "seconds")
    print("Estimated center:", (res[-1][0],res[-1][1],res[-1][2]))
    return (res,liste_decr)
```

## Moving particles

### Formulation using coordinates vectors

```python
def Create_A_and_b_Moving_Patricle(W):
    "Cette fonction permet de creer la matrice A de taille n * 5m et le vecteur b de taille n"
    
    #Convertir la matrice W en 2D -- Taille: n * s(3) avec n = s(1)*s(2)
    s1 = np.size(W,0)
    s2 = np.size(W,1)
    s3 = np.size(W,2)
    
    W_2D = W.reshape(s1*s2, s3)

    
    # Definir les vecteurs x et y de taille n chacun
    x = np.zeros((s1*s2,1))
    y = np.zeros((s1*s2,1))
    for jj in range(s1*s2):
        x[jj] = np.floor((jj)/s1)
        y[jj] = np.mod(jj,s1)
  
# Definir un vecteur unite de taille n
    vecteur_unitaire = np.ones((s1*s2,1));    
        
    A = np.zeros((np.size(W_2D, 0), 5*s3, s3))
    b = np.zeros((np.size(W_2D, 0), s3));
    
    listA = [] #empty listA
    listb = []  #empty listb
    
    for kk in range(s3):
        #Definir le vecteur f_i de taille 5m (vecteur de 0 sauf le (i)th element est 1, avec i = 1, ... m)    
        f_i = np.zeros((5*s3,1))

        #Definir le vecteur f_m_i de taille 5m (vecteur de 0 sauf le (m+i)th element est 1, avec i = 1, ... m)
        f_m_i = np.zeros((5*s3,1))    

        #Definir le vecteur f_3m_i de taille 5m (vecteur de 0 sauf le (3m+i)th element est 1, avec i=1, ... m)
        f_3m_i = np.zeros((5*s3,1))

        #Definir le vecteur f_4m_i de taille 5m (vecteur de 0 sauf le (4m+i)th element est 1, avec i=1, ... m)
        f_4m_i = np.zeros((5*s3,1))  
        
        f_i[kk] = 1
        f_m_i[s3+kk] = 1
        f_3m_i[3*s3+kk] = 1
        f_4m_i[4*s3+kk] = 1
        
        A[:, :, kk] = np.diag(np.sqrt(W_2D[:, kk])) .dot(-2*np.kron(np.transpose(f_i), x) - 2*np.kron(np.transpose(f_m_i), y) + np.kron(np.transpose(f_3m_i), vecteur_unitaire) - np.kron(np.transpose(f_4m_i), vecteur_unitaire))
        listA.append(A[:, :, kk])
        
        b[:, kk] = np.multiply(-np.sqrt(W_2D[:, kk]), np.transpose(x**2)) + np.multiply(-np.sqrt(W_2D[:, kk]), np.transpose(y**2))
        listb.append(b[:, kk])
               
    
    return listA, listb
```

### IP with Newton, moving particle version (with regularization)

```python
import numpy as np
import matplotlib.pyplot as plt

# Centering step

# fonction à minimiser
def fct(listA, listb, lambd, lambd_z):
    m = len(listA)
    def f(v): 
        reg = np.sum((v[1:m]-v[0:m-1])**2) + np.sum((v[m+1:2*m]-v[m:2*m-1])**2) # regularize on the x and y axis
        reg_z = np.sum((v[2*m+1:3*m]-v[2*m:3*m-1])**2)  # regularize on the z-axis 
        return np.sum([np.linalg.norm(listA[i]@v - listb[i])**2 for i in range(m)], axis=0)/2 + lambd*reg +lambd_z*reg_z
    return f

# v = np.concatenate((alpha, beta, gamma, xsi, eta))


# Fonction avec barrière:
def fct_bar(listA, listb, z, R, d, t, lambd, lambd_z): # t est le paramètre de barrière
    m = len(listA)
    def f(v):
        constr1 = v[4*m:]
        constr2 = -(v[4*m:] + (v[2*m:3*m]-z)**2 - R**2)
        constr3 = -(v[0:m]**2 + v[1*m:2*m]**2 - v[3*m:4*m])
        constr4 = -(v[3*m:4*m] - d**2)
        constr = np.concatenate((constr1,constr2,constr3,constr4))
        reg = np.sum((v[1:m]-v[0:m-1])**2) + np.sum((v[m+1:2*m]-v[m:2*m-1])**2) 
        reg_z = np.sum((v[2*m+1:3*m]-v[2*m:3*m-1])**2)
        if np.min(constr) <= 0:
            #raise NameError("v non admissible")
            return np.inf
        else:
            return np.sum([np.linalg.norm(listA[i]@v - listb[i])**2 for i in range(m)])/2 - (1/t)*np.sum(np.log(constr), axis=0) + lambd*reg + lambd_z*reg_z
    return f



# Gradient en v de la fonction à minimiser
def grad_f(listA, listb, z, R, d, t, v, lambd, lambd_z):
    n=len(v)
    m = len(z)
    res = np.sum([listA[i].T@(listA[i]@v - listb[i]) for i in range(m)], axis=0)
    bar = np.zeros(n).astype(float)
    # terme dû à la fonction barrière pour constr3 et constr4
    x = (-(v[0:m]**2 + v[1*m:2*m]**2 - v[3*m:4*m]))
    bar[0:m] += -2*v[0:m]/x
    bar[1*m:2*m] += -2*v[1*m:2*m]/x
    bar[3*m:4*m] += 1/x - 1/(d**2-v[3*m:4*m])
    x = R**2 - v[4*m:] - (v[2*m:3*m]-z)**2
    bar[2*m:3*m] += 2*(z-v[2*m:3*m])/x
    bar[4*m:] += 1/v[4*m:] - 1/x 
    # terme dû à la regularisation
    reggrad = np.zeros(n).astype(float)
    reggrad[0] = -2*(v[1]-v[0])*lambd
    reggrad[m-1] = 2*(v[m-1]-v[m-2])*lambd
    reggrad[m] = -2*(v[m+1]-v[m])*lambd
    reggrad[2*m-1] = 2*(v[2*m-1]-v[2*m-2])*lambd
    reggrad[2*m] = -2*(v[2*m+1]-v[2*m])*lambd_z
    reggrad[3*m-1] = 2*(v[3*m-1]-v[3*m-2])*lambd_z
    for i in range(1,m-1):
        reggrad[i] = -2*(v[i+1]-v[i]) + 2*(v[i]-v[i-1])*lambd  
        reggrad[m+i] = -2*(v[m+i+1]-v[m+i]) + 2*(v[m+i]-v[m+i-1])*lambd 
        reggrad[2*m+i] = -2*(v[2*m+i+1]-v[2*m+i]) + 2*(v[2*m+i]-v[2*m+i-1])*lambd_z
    return res - bar/t + reggrad



# Hessienne
def hess_f(listA, listb, z, R, d, t, v, lambd, lambd_z):
    n=len(v)
    m = len(z)
    res = np.sum([listA[i].T@listA[i] for i in range(m)], axis=0)
    bar = np.zeros((n,n)).astype(float)
    for i in range(m):
        x = (-(v[i]**2 + v[m+i]**2 - v[3*m+i]))**2
        y = (R**2 - v[4*m+i] - (v[2*m+i]-z[i])**2)**2
        bar[i,i] += (-2*(v[3*m+i] - v[i]**2 - v[1*m+i]**2) - 4*v[i]**2)/x
        bar[m+i,m+i] += (-2*(v[3*m+i] - v[i]**2 - v[1*m+i]**2) - 4*v[1*m+i]**2)/x
        bar[2*m+i,2*m+i] += (-2*(R**2 - v[4*m+i] - (v[2*m+i]-z[i])**2) - 4*(v[2*m+i]-z[i])**2)/y
        bar[3*m+i,3*m+i] += -1/x - 1/(v[3*m+i]-d**2)**2
        bar[4*m+i,4*m+i] += -1/v[4*m+i]**2 - 1/y
        b = -4*v[i]*v[m+i]/x
        bar[i,m+i] += b
        bar[m+i,i] += b
        b = 2*v[i]/x
        bar[i,3*m+i] += b
        bar[3*m+i,i] += b
        b = 2*v[m+i]/x
        bar[m+i,3*m+i] += b
        bar[3*m+i,m+i] += b
        b = 2*(v[2*m+i]-z[i])/y
        bar[2*m+i,4*m+i] += b
        bar[4*m+i,2*m+i] += b
    # terme dû à la regularisation
    reghess = 4*np.diag([1 if i<(3*m) else 0 for i in range(n)]).astype(float)# les termes diagnonaux valent 2, 4 ou 0                                                                        
    reghess[0,0] = 2 
    reghess[m-1,m-1] = 2
    reghess[m,m] = 2
    reghess[2*m-1,2*m+1] = 2
    reghess[2*m,2*m] = 2
    reghess[3*m-1,3*m-1] = 2 
    for i in range(1,m-1):
        reghess[i,i-1] = -2
        reghess[i-1,i] = -2                                                                         
        reghess[i,i+1] = -2                                                                         
        reghess[i+1,i] = -2                                                                         
        reghess[m+i,m+i-1] = -2
        reghess[m+i-1,m+i] = -2 
        reghess[m+i+1,m+i] = -2
        reghess[m+i,m+i+1] = -2                                                                         
        reghess[2*m+i,2*m+i-1] = -2
        reghess[2*m+i-1,2*m+i] = -2                                                                         
        reghess[2*m+i+1,2*m+i] = -2
        reghess[2*m+i,2*m+i+1] = -2   
    reghess[0:2*m, 0:2*m] = lambd*reghess[0:2*m, 0:2*m]
    reghess[2*m:3*m, 2*m:3*m] = lambd_z*reghess[2*m:3*m, 2*m:3*m]
    return  res - bar/t + reghess





# Backtracking line search
def lineSearch(f,x,delta,grad,alpha,beta,t=1): # 0<alpha<1/2 et 0<beta<1
    while f(x+t*delta) >= f(x) + alpha*t*(grad.T@delta) :
        t = beta*t
    return t

def graddesc(f,v,grad, t=1000):
    crit = True
    while crit:
        if f(v - t*grad) < f(v):
            v = v - t*grad
            crit = False
        t = t/2
    return v

# Centering step
# On retourne la liste des points et des decr pour les graphes à la fin
def centering_step(listA, listb, z, R, d, t, v0, crit, alpha = 0.4, beta = 0.9, lambd=0, lambd_z=0): # t et eps > 0
    # méthode de Newton
    res = [v0]
    v = v0
    decr = np.inf
    liste_decr=[]
    f = fct_bar(listA, listb, z, R, d, t,lambd,lambd_z)
    grad = grad_f(listA, listb, z, R, d, t, v, lambd,lambd_z)
    hess = hess_f(listA, listb, z, R, d, t, v, lambd,lambd_z)
    hess_inv = np.linalg.inv(hess)
    delta = -hess_inv@grad
    decr = grad.T@hess_inv@grad
    #print("decr = ",decr)
    #print(np.linalg.norm(grad))
    #while np.linalg.norm(grad) > crit:
    while decr/2 >= crit:
        v = v + lineSearch(f,v,delta,grad,alpha,beta)*delta
        #v = graddesc(f,v,grad)
        res.append(v)
        grad = grad_f(listA, listb, z, R, d, t, v, lambd,lambd_z)
        hess = hess_f(listA, listb, z, R, d, t, v, lambd,lambd_z)
        hess_inv = np.linalg.inv(hess)
        delta = -hess_inv@grad
        decr = grad.T@hess_inv@grad
        liste_decr.append(decr)
        print("decr = ",decr)
        #print(delta)
        #print("v_cent = ",v)
        #print("f_bar(v_cent) = ",f(v))
    return (res,liste_decr)


# Barrier method
# On retourne la liste des points et des decr pour les graphes à la fin
def barr_method(listA, listb, z, R, d, v0,eps, crit, alpha =0.4, beta = 0.9, t0 = 1, mu = 10, lambd=0, lambd_z=0): # t0, mu et eps > 0
    start = time.time()
    res = [v0]
    liste_decr = []
    n = len(v0)
    v = v0
    t = t0
    while n/t >= eps:
        #print("v = ",v)
        print("f(v) = ",fct(listA,listb,lambd,lambd_z)(v), " Barrier parameter = ", t)
        (v, liste_decr_centering) = centering_step(listA, listb, z, R, d, t, v,crit, alpha, beta, lambd,lambd_z)
        liste_decr+= liste_decr_centering
        t = mu*t
        res = res + v
        v = v[-1]
    print("Computation time: ", time.time()-start, "seconds")
    print("Mean estimated center:", (np.mean(res[-1][0:m]),np.mean(res[-1][m:2*m]),np.mean(res[-1][2*m:3*m])))
    return (res,liste_decr)

 
        
```
